Let's see which variables have 5 years of data. We will select only those variables with enough time data to see possible correlations and trends.

```python
def plot_cat_time(data, var, year, plot=True, ax=None, categorical=True, bins=2, labels=['low', 'high']):
    """Plots categorical variable var from data against year.
    
    data - dataframe with variables var and year.
    var - column name from data to plot against time
    year - column name from data indicating year
    plot - if False, just returns the dataframe used for plotting
    ax - matplotlib axes to plot on if plot is True
    categorical - If variable is not categorical, function uses bins and labels
        to convert variable into a categorical
    """
    if (categorical):
        temp = pd.crosstab(data[year], data[var]) \
                  .drop('unknown', axis=1, errors='ignore') \
                  .drop('unknown', axis=0, errors='ignore') \
                  .apply(lambda x: x / x.sum(), 1)
        if plot: temp.plot.bar(stacked=True, ax=ax)
    else: #convert numerical into categorical
        temp = pd.crosstab(data[year], pd.cut(data[var], bins=bins, labels=labels)) \
                  .drop('unknown', axis=1, errors='ignore') \
                  .drop('unknown', axis=0, errors='ignore') \
                  .apply(lambda x: x / x.sum(), 1)
        if plot: temp.plot.bar(stacked=True, ax=ax)
        
    if plot:
        ax.set_xlabel("")
        ax.set_title(var)
        ax.legend(bbox_to_anchor=(1.05, .5, .2, .5), loc='upper right',
                  fontsize='small', borderaxespad=0.1, fancybox=True, shadow=True)
        
    return temp
```

```python
cat_var = x_var + target_cat_var #aggregate all categorical variables
fig, ax = plt.subplots(len(cat_var) + len(target_num_var),1, figsize=(15,60)) #make enough plots

#plot all variables against time
for i in range(len(cat_var) + len(target_num_var)):
    if (i<len(cat_var)): #if categorical
        plot_cat_time(df, cat_var[i], 'year', ax=ax[i])
    else: #if numerical, then call the part of teh plot_cat_time function that bins the variable
        plot_cat_time(df, target_num_var[i-len(cat_var)], 'year', ax=ax[i], categorical=False, 
                        bins=4, labels=['very low', 'low', 'medium', 'high'])
    
fig.suptitle('Variables Over 5 Years', y=.98)
fig.tight_layout()
fig.subplots_adjust(top=.97)
fname = 'figures/variables_13_17.png'
fig.savefig(fname, format='png')
```

```python
def plot_num_time(data, var, year, plot=True, ax=None, percent=False):
    """Plots numerical variable var from data against year if plot is true. 
    Returns dataframe temp, which has descriptive statistics of var over years.
    
    data - dataframe with variables var and year.
    var - column name from data to plot against time
    year - column name from data indicating year
    plot - if false, then function only returns dataframe used for plotting
    ax - matplotlib axes to plot on if plot is true
    percent - if True, plots the percent difference from the first year
    """
    if (percent==False):
        temp = data.loc[:, [var, year]] \
                  .groupby(year) \
                  .agg(['mean', 'std', 'count']) 

        temp.columns = ['mean', 'std', 'count'] #remove multi-index
        temp['std_err'] = temp['std'] / np.sqrt(temp['count'])
        if plot: temp['mean'].plot(kind='line', ax=ax, yerr=temp.std_err, ecolor='gray')
    else:
        temp = data.loc[:, [year, var]] \
                 .groupby(year) \
                 .agg('mean')
        temp.columns = ['mean']
        temp['first_year_mean'] = temp.loc[temp.index[0],'mean']
        temp['percent_diff'] = temp[['first_year_mean', 'mean']].pct_change(axis='columns')['mean']
        if plot: 
            temp['percent_diff'].plot(kind='line',ax=ax)
            ax.get_yaxis().set_major_formatter(
                mpl.ticker.FuncFormatter(lambda y,p: str(int(y*100))+'%'))   #format the y-axis to percentages
    
    if plot:
        ax.set_xlabel("")
        ax.set_title(var)
    
    return temp
```

```python
#plot the 4 target variables over time independent of any sub-groups
fig, ax = plt.subplots(2,2,figsize=(12,12))

#convert smoke_now into a percentage for the value "every_day" across years
smoke_stats = helper.analyze_cat_cat(df, 'year', 'smoke_now')['percent'].unstack()['every_day']
smoke_stats.pct_change().fillna(0).plot(kind='line',ax=ax[0,0]) #smoking
ax[0,0].get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda y,p: str(int(y*100))+'%'))   #format the y-axis to percentages

plot_num_time(df.loc[:, ['avg_drinks', 'year']], 'avg_drinks', 'year', ax=ax[0,1], percent=True) #avg_drinks
plot_num_time(df.loc[:, ['many_drinks', 'year']], 'many_drinks', 'year', ax=ax[1,0], percent=True) #many_drinks
plot_num_time(df.loc[:, ['max_drinks', 'year']], 'max_drinks', 'year', ax=ax[1,1], percent=True) #max_drinks

#bit of formatting, and taking 2018 out since it has a really small sample size comparatively
ax[0,0].set_xlabel("")
ax[0,0].set_title('Smoke Every Day')
ax[0,0].set_xlim(2013, 2017)
ax[0,1].set_xlim(2013, 2017)
ax[1,0].set_xlim(2013, 2017)
ax[1,1].set_xlim(2013, 2017)

ax[0,0].set_ylim(-.1, .1)
ax[0,1].set_ylim(-.1, .1)
ax[1,0].set_ylim(-.1, .1)
ax[1,1].set_ylim(-.1, .1)


fig.suptitle('Percent Change Target Variables Since 2013')
fname = 'figures/target_variables_13_17.png'
fig.savefig(fname, format='png')
```

Ok, so down and up, but not by much and hard to draw a correlation. Now let's see if these trends hold true across all states, or if there are any states that have some interesting trends. We'll narrow this analysis to smoke_now and many_drinks, which seemed to have the best separation across environmental variables from previous analyses.

```python
fig,ax=plt.subplots(1,1,figsize=(16,12))

#create a dataframe of percent changes in each target variable from 2013
states = df.state.dropna().unique()
smoke_changes = pd.DataFrame(index=states, columns=['change'])
for state in states:
    smoke_stats = helper.analyze_cat_cat(df.loc[df.state==state,:], 'year', 'smoke_now')['percent'].unstack()
    smoke_stats = smoke_stats['every_day']
    smoke_changes.loc[state,'change'] = (smoke_stats[2017] - smoke_stats[2013]) / smoke_stats[2013]
    
smoke_changes.loc[:,'change'].sort_values().plot.bar(ax=ax)
ax.set_title('Percent Change in Proportion of Every Day Smokers from 2013 to 2017')
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda y,p: str(int(y*100))+'%'))   #format the y-axis to percentages
```

```python
fig,ax=plt.subplots(1,1,figsize=(16,12))

#create a dataframe of percent changes in each target variable from 2013
many_drink_changes = pd.DataFrame(index=states, columns=['change'])
for state in states:
    many_drink_stats = plot_num_time(df.loc[df.state==state, ['many_drinks', 'year']], 
                                    'many_drinks', 'year', percent=True, plot=False) #avg_drinks
    many_drink_changes.loc[state,'change'] = many_drink_stats.loc[2017,'percent_diff']

many_drink_changes.loc[:,'change'].sort_values().plot.bar(ax=ax)
ax.set_title('Percent Change in Number of Occasions of Many Drinks Per Person from 2013 to 2017')
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda y,p: str(int(y*100))+'%'))   #format the y-axis to percentages
```

```python
fig,ax=plt.subplots(1,1,figsize=(16,12))

#create a dataframe of percent changes in each target variable from 2013
dep_changes = pd.DataFrame(index=states, columns=['change'])
for state in states:
    dep_stats = helper.analyze_cat_cat(df.loc[df.state==state,:], 'year', 'depression')['percent'].unstack()
    dep_stats = dep_stats['yes']
    dep_changes.loc[state,'change'] = (dep_stats[2017] - dep_stats[2013]) / dep_stats[2013]
    
dep_changes.loc[:,'change'].sort_values().plot.bar(ax=ax)
ax.set_title('Percent Change in Proportion Indicating Depression from 2013 to 2017')
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda y,p: str(int(y*100))+'%'))   #format the y-axis to percentages
```

```python
fig,ax=plt.subplots(1,1,figsize=(16,12))

#create a dataframe of percent changes in each target variable from 2013
mh_changes = pd.DataFrame(index=states, columns=['change'])
for state in states:
    mh_stats = helper.analyze_cat_cat(df.loc[df.state==state,:], 'year', 'poor_mental_health_days')['percent'].unstack()
    mh_stats = (mh_stats['<20'] + mh_stats['>20'])
    mh_changes.loc[state,'change'] = (mh_stats[2017] - mh_stats[2013]) / mh_stats[2013]
    
mh_changes.loc[:,'change'].sort_values().plot.bar(ax=ax)
ax.set_title('Percent Change in Proportion Indicating >10 Poor Mental Health Days from 2013 to 2017')
ax.get_yaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda y,p: str(int(y*100))+'%'))   #format the y-axis to percentages
```

We'll grab the top and bottom states for smoking and alcohol. 

For smoking, we'll look at 

- Decrease: (New York, Massachusetts, Alabama, New Jersey, California) and compare to
- Increase: (South Dakota, Pennsylvania, Louisiana, and Florida), we'll leave Guam out due to sample size

For alcohol, we'll look at

We're interested in whether the changes in substance use in these states is correlated with changes in environmental variables, including satisfied life, depression prevalence, and number of poor mental health days. These environmental variables had the greatest separation in previous analyses.

```python
#create lists for each group we want to test
smoking_inc_group = ['DC', 'LA', 'PR', 'FL']
smoking_dec_group = ['MA', 'CT', 'OK', 'AR']
alcohol_inc_group = ['TN', 'WY', 'KY', 'LA']
alcohol_dec_group = ['CT', 'SD', 'OK', 'IL']
var = ['smoke_now', 'many_drinks', 'year', 'sex', 'age', 'marital', 'education', 'employment', 'income', 'emotional_support', 
         'satisfied_life', 'depression', 'poor_mental_health_days']

#get data for target and environmental variables for each state in the groups
smoking_inc_data = df.loc[df.state_code.isin(smoking_inc_group), var]
smoking_dec_data = df.loc[df.state_code.isin(smoking_dec_group), var]
alcohol_inc_data = df.loc[df.state_code.isin(alcohol_inc_group), var]
alcohol_dec_data = df.loc[df.state_code.isin(alcohol_dec_group), var]
```

```python
#get smoking stats for inc and dec group
temp = plot_cat_time(smoking_inc_data, 'smoke_now', 'year', plot=False).drop(['not_at_all', 'some_days'], axis=1)
temp['first_year'] = temp.loc[2013,'every_day']
smoking_inc = temp[['first_year', 'every_day']].pct_change(axis='columns')['every_day']

temp = plot_cat_time(smoking_dec_data, 'smoke_now', 'year', plot=False).drop(['not_at_all', 'some_days'], axis=1)
temp['first_year'] = temp.loc[2013,'every_day']
smoking_dec = temp[['first_year', 'every_day']].pct_change(axis='columns')['every_day']

#get percentage of pop reporting >10 days poor mental health for each smoking group
temp = plot_cat_time(smoking_inc_data, 'poor_mental_health_days', 'year', plot=False)
temp = (temp['<20']+temp['>20']).to_frame().rename({0:"mh"},axis=1)
temp['first_year'] = temp.loc[2013,'mh']
smoking_inc_mh = temp[['first_year', 'mh']].pct_change(axis='columns')['mh']

temp = plot_cat_time(smoking_dec_data, 'poor_mental_health_days', 'year', plot=False)
temp = (temp['<20']+temp['>20']).to_frame().rename({0:"mh"},axis=1)
temp['first_year'] = temp.loc[2013,'mh']
smoking_dec_mh = temp[['first_year', 'mh']].pct_change(axis='columns')['mh']

#get percentage of pop reporting depression for each smoking group
temp = plot_cat_time(smoking_inc_data, 'depression', 'year', plot=False)
temp = (temp['yes']).to_frame()
temp['first_year'] = temp.loc[2013,'yes']
smoking_inc_dep = temp[['first_year', 'yes']].pct_change(axis='columns')['yes']

temp = plot_cat_time(smoking_dec_data, 'depression', 'year', plot=False)
temp = (temp['yes']).to_frame()
temp['first_year'] = temp.loc[2013,'yes']
smoking_dec_dep = temp[['first_year', 'yes']].pct_change(axis='columns')['yes']

#generate plots
fig, ax = plt.subplots(2,2,figsize=(16,9))
ax[0,0].plot(smoking_inc, label='Top 5 States')
ax[0,0].plot(smoking_dec, label='Bottom 5 States')
ax[0,0].set_title('Percent Change in Proportion of Every Day Smokers from 2013')
ax[0,0].set_xlim(2013,2017)
ax[0,0].legend()

ax[0,1].plot(smoking_inc_mh, label='Top 5 States')
ax[0,1].plot(smoking_dec_mh, label='Bottom 5 States')
ax[0,1].set_title('Percent Change in Proportion with >10 Poor Mental Health Days from 2013')
ax[0,1].set_xlim(2013,2017)
ax[0,1].legend()

ax[1,1].plot(smoking_inc_dep, label='Top 5 States')
ax[1,1].plot(smoking_dec_dep, label='Bottom 5 States')
ax[1,1].set_title('Percent Change in Proportion Indicating Depression from 2013')
ax[1,1].set_xlim(2013,2017)
ax[1,1].legend()
```

```python
#get alcohol stats for inc and dec group
alcohol_inc = plot_num_time(alcohol_inc_data, 'many_drinks', 'year', plot=False, percent=True)['percent_diff']

alcohol_dec = plot_num_time(alcohol_dec_data, 'many_drinks', 'year', plot=False, percent=True)['percent_diff']

#get percentage of pop reporting >10 days poor mental health for each alcohol group
temp = plot_cat_time(alcohol_inc_data, 'poor_mental_health_days', 'year', plot=False)
temp = (temp['<20']+temp['>20']).to_frame().rename({0:"mh"},axis=1)
temp['first_year'] = temp.loc[2013,'mh']
alcohol_inc_mh = temp[['first_year', 'mh']].pct_change(axis='columns')['mh']

temp = plot_cat_time(alcohol_dec_data, 'poor_mental_health_days', 'year', plot=False)
temp = (temp['<20']+temp['>20']).to_frame().rename({0:"mh"},axis=1)
temp['first_year'] = temp.loc[2013,'mh']
alcohol_dec_mh = temp[['first_year', 'mh']].pct_change(axis='columns')['mh']

#get percentage of pop reporting depression for each alcohol group
temp = plot_cat_time(alcohol_inc_data, 'depression', 'year', plot=False)
temp = (temp['yes']).to_frame()
temp['first_year'] = temp.loc[2013,'yes']
alcohol_inc_dep = temp[['first_year', 'yes']].pct_change(axis='columns')['yes']

temp = plot_cat_time(alcohol_dec_data, 'depression', 'year', plot=False)
temp = (temp['yes']).to_frame()
temp['first_year'] = temp.loc[2013,'yes']
alcohol_dec_dep = temp[['first_year', 'yes']].pct_change(axis='columns')['yes']

fig, ax = plt.subplots(2,2,figsize=(16,9))
ax[0,0].plot(alcohol_inc, label='Top 5 States')
ax[0,0].plot(alcohol_dec, label='Bottom 5 States')
ax[0,0].set_title('Percent Change in Number of Binge Drinking Events Per 30 Days from 2013')
ax[0,0].set_xlim(2013,2017)
ax[0,0].legend()

ax[0,1].plot(alcohol_inc_mh, label='Top 5 States')
ax[0,1].plot(alcohol_dec_mh, label='Bottom 5 States')
ax[0,1].set_title('Proportion with >10 Poor Mental Health Days')
ax[0,1].set_xlim(2013,2017)
ax[0,1].legend()

ax[1,1].plot(alcohol_inc_dep, label='Top 5 States')
ax[1,1].plot(alcohol_dec_dep, label='Bottom 5 States')
ax[1,1].set_title('Proportion Indicating Depression')
ax[1,1].set_xlim(2013,2017)
ax[1,1].legend()
```
